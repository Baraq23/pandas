# E-commerce purchases

```python
import pandas as pd
url = "https://learn.zone01kisumu.ke/api/content/root/public/subjects/ai/pandas/data/Ecommerce_purchases.txt"
```

```python
df = pd.read_csv(url)
print(df.shape)

averagePurchasePrice = df['Purchase Price'].mean()
print("average Purchase price:",averagePurchasePrice)

highPurchasePrice = df['Purchase Price'].max()
lowPurchasePrice = df['Purchase Price'].min()
print("Highest Purchase Price:", highPurchasePrice)
print("Lowest Purchase Price:", lowPurchasePrice)

print("People with English as language of choice:", df[df['Language'] == 'en'].shape[0])


print("People who are lawyers:", df[df['Job'] == 'Lawyer'].shape[0])

print("Purchases made during AM and PM:", df['AM or PM'].value_counts())

print("The 5 most common job titles:", df['Job'].value_counts().head(5))

print("Purchase price of Lot 90 WT:", df[df['Lot'] == '90 WT']['Purchase Price'])

email  =  df[df['Credit Card'] == 4926535242672853]['Email']
print("Email of person with Credit card Number 4926535242672853:", email)

peopleWithExpress = df[(df['CC Provider'] == 'American Express') & (df['Purchase Price'] > 95)].shape[0]
print("People with American express as credit card provider:", peopleWithExpress)

creditExpires25 = df[df['CC Exp Date'].apply(lambda x: x.split('/')[1] == '25')].shape[0]
print("Credit Expires in 2025", creditExpires25)

popularEmail = df['Email'].apply(lambda x: x.split('@')[1]).value_counts().head(5)
print("Most popular email providers:", popularEmail)
```
